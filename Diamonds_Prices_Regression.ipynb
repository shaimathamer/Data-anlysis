```python
diamonds['volume'] = diamonds['x'] * diamonds['y'] * diamonds['z']
diamonds = diamonds.drop(['x', 'y', 'z'], axis=1)

sns.scatterplot(x='volume', y='price', data=diamonds)
plt.show()
```

```python
diamonds.replace({'Ideal': 5, 'Premium': 4, 'Very Good': 3, 'Good': 2, 'Fair': 1}, inplace=True)
diamonds.replace({'D': 7, 'E': 6, 'F': 5, 'G': 4, 'H': 3, 'I': 2, 'J': 1}, inplace=True)
diamonds.replace({'IF': 8, 'VVS1': 7, 'VVS2': 6, 'VS1': 5, 'VS2': 4, 'SI1': 3, 'SI2': 2, 'I1': 1}, inplace=True)

from scipy import stats

diamonds = diamonds[(np.abs(stats.zscore(diamonds)) < 3).all(axis=1)]
```

```python
sns.scatterplot(x='volume', y='price', data=diamonds)
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

std_scaler = StandardScaler()
numerical_attributes = pd.DataFrame(
    std_scaler.fit_transform(diamonds[['carat', 'cut', 'clarity', 'depth', 'table', 'volume']]), diamonds.index,
    ['carat', 'cut', 'clarity', 'depth', 'table', 'volume'])
diamonds_std = diamonds.copy()
diamonds_std[['carat', 'cut', 'clarity', 'depth', 'table', 'volume']] = numerical_attributes[
    ['carat', 'cut', 'clarity', 'depth', 'table', 'volume']]
```

```python
diamonds.corr()['price'].sort_values(ascending=False)
```

```python
features = diamonds_std.drop(['price'], axis=1)
labels = diamonds_std.price
from sklearn.model_selection import train_test_split

train_features, test_features, train_labels, test_labels = train_test_split(features, labels, random_state=42,
                                                                            test_size=0.2)
```

```python
from sklearn import linear_model
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

lin_reg = linear_model.LinearRegression()
lin_reg.fit(train_features, train_labels)
lin_reg_test_predicted = lin_reg.predict(test_features)
lin_reg_scores = cross_val_score(lin_reg, train_features, train_labels, scoring='neg_mean_squared_error', cv=10)
print('Linear regression:')
print('Cross-validation scores:', np.sqrt(-lin_reg_scores))
print('Cross-validation scores average:', np.average(np.sqrt(-lin_reg_scores)))
print('Cross-validation scores standard deviation:', np.sqrt(-lin_reg_scores).std())
lin_reg_mse = mean_squared_error(test_labels, lin_reg_test_predicted)
lin_reg_rmse = mean_squared_error(test_labels, lin_reg_test_predicted) ** 0.5
lin_reg_r2 = r2_score(test_labels, lin_reg_test_predicted)
print('MSE:', lin_reg_mse)
print('RMSE:', lin_reg_rmse)
print('R2:', lin_reg_r2)
```

```python
from yellowbrick.regressor import PredictionError

visualizer = PredictionError(lin_reg)
visualizer.fit(train_features, train_labels)
visualizer.score(test_features, test_labels)
visualizer.show()
```

```python
from sklearn.tree import DecisionTreeRegressor

tree_reg = DecisionTreeRegressor()
tree_reg.fit(train_features, train_labels)
tree_reg_test_predicted = tree_reg.predict(test_features)
tree_reg_scores = cross_val_score(tree_reg, train_features, train_labels, scoring='neg_mean_squared_error', cv=10)
print('Decision tree:')
print('Cross-validation scores:', np.sqrt(-tree_reg_scores))
print('Cross-validation scores average:', np.average(np.sqrt(-tree_reg_scores)))
print('Cross-validation scores standard deviation:', np.sqrt(-tree_reg_scores).std())
tree_reg_mse = mean_squared_error(test_labels, tree_reg_test_predicted)
tree_reg_rmse = mean_squared_error(test_labels, tree_reg_test_predicted) ** 0.5
tree_reg_r2 = r2_score(test_labels, tree_reg_test_predicted)
print('MSE:', tree_reg_mse)
print('RMSE:', tree_reg_rmse)
print('R2:', tree_reg_r2)
```

```python
visualizer = PredictionError(tree_reg)
visualizer.fit(train_features, train_labels)
visualizer.score(test_features, test_labels)
visualizer.show()
```

```python
from sklearn.ensemble import RandomForestRegressor

forest_reg = RandomForestRegressor()
forest_reg.fit(train_features, train_labels)
forest_reg_test_predicted = forest_reg.predict(test_features)
forest_reg_scores = cross_val_score(forest_reg, train_features, train_labels, scoring='neg_mean_squared_error', cv=10)
print('Random forest:')
print('Cross-validation scores:', np.sqrt(-forest_reg_scores))
print('Cross-validation scores average:', np.average(np.sqrt(-forest_reg_scores)))
print('Cross-validation scores standard deviation:', np.sqrt(-forest_reg_scores).std())
forest_reg_mse = mean_squared_error(test_labels, forest_reg_test_predicted)
forest_reg_rmse = mean_squared_error(test_labels, forest_reg_test_predicted) ** 0.5
forest_reg_r2 = r2_score(test_labels, forest_reg_test_predicted)
print('MSE:', forest_reg_mse)
print('RMSE:', forest_reg_rmse)
print('R2:', forest_reg_r2)
```

```python
visualizer = PredictionError(forest_reg)
visualizer.fit(train_features, train_labels)
visualizer.score(test_features, test_labels)
visualizer.show()
```
